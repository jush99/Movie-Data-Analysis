```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize']=(10,8)## Adjusting the configuration of plots we create

```

```python
df=pd.read_csv('movies.csv')
```

```python
df
```

```python
df=df.fillna(0)
```

```python
#Looking for missing data
for i in df.columns:
    pct_missing=np.mean(df[i].isnull())
    print('{}-{}%'.format(i,pct_missing))
```

```python
df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
```

```python
df.dtypes
```

```python
df=df.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
pd.set_option('display.max_rows',30)
```

```python
# Dropping duplicates
#df['company'].drop_duplicates().sort_values(ascending=False)
df.drop_duplicates()
```

```python
# To compare budget and gross revenue we use the scatter plot

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
df.head()
```

```python
# Plot the budget vs gross to find correlation.
#Here we see the budget and gross are corelated we do not how much but the slope is increrasing in positive direction
sns.regplot(x='budget',y='gross',data=df,scatter_kws={'color':'red'},line_kws={'color':'blue'})
```

```python
df.corr(method='pearson')# Method can be pearson,kendall,spearman.If nothing is mentioned it will be pearson
```

```python
correlation_matrix=df.corr(method='pearson')
plt.title('Correlation between Movie features')

sns.heatmap(correlation_matrix,annot=True)
plt.show()
```

```python
df_numerized=df
for i in df_numerized.columns:
    if(df_numerized[i].dtype=='object'):
        df_numerized[i]=df_numerized[i].astype('category')
        df_numerized[i]=df_numerized[i].cat.codes
df_numerized        
```

```python
correlation_matrix=df_numerized.corr(method='pearson')
plt.title('Correlation between Movie features')

sns.heatmap(correlation_matrix,annot=True)
plt.show()
```

```python
correlation_mat=df_numerized.corr()
corr_pairs=correlation_mat.unstack()
sorted_pairs=corr_pairs.sort_values()
high_corr=sorted_pairs[(sorted_pairs)>0.5]
high_corr
```
